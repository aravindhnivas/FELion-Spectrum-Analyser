```python
from symfit import variables, Parameter, ODEModel, D, Fit, parameters
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = 'kineticfile.txt'
data = np.genfromtxt(file)
t = data[:, 0]
CD = data[:, 1]
CDHe = data[:, 3]
CDHe2 = data[:, 4]
CDHe3 = data[:, 5]
CDHe4 = data[:, 6]
He = 1.768e14

total = CD+CDHe+CDHe2+CDHe3+CDHe4
```

```python
fig, ax = plt.subplots(figsize = (7, 4), dpi = 150)

ax.plot(t, total,'k', label = 'Total')
ax.plot(t, CD, lw=1, label = 'CD+')
ax.plot(t, CDHe, lw=1, label = 'CDHe+')
ax.plot(t, CDHe2, lw=1, label = 'CDHe2+')
ax.plot(t, CDHe3, lw=1, label = 'CDHe3+')

ax.set_xlim(left = 0)
ax.set_yscale('log')
ax.legend()
ax.grid()

plt.show()
plt.close()
```
