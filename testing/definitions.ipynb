```python
from tkinter import *
from tkinter import ttk, StringVar, IntVar

class my_label(Frame):
    def __init__(self, parent, label, font):
        Frame.__init__(self, parent)
        
        self.label = label
        self.font = font

        self.label = Label(self, text=self.label, font = self.font)
        self.label.place(relx = 0, rely = 0, relwidth = 1, relheight = 1)

class my_entry(Frame):
    def __init__(self, parent, entry, font):
        Frame.__init__(self, parent)

        self.entry = entry

        if self.entry is int(): self.value = IntVar()
        else: self.value = StringVar()
        
        self.value.set(self.entry)       
        self.entry = Entry(self, textvariable = self.value, bg = "white",\
                              bd = 5, font = font)

        self.entry.place(relx = 0, rely = 0, relwidth = 1, relheight = 1)
        
        self.get()
        
    def get(self):
        return self.value.get()

class my_button(Frame):
    
    def __init__(self, parent, name, func, parameter):
        Frame.__init__(self, parent)
        
        self.name = name
        self.func = func
        self.parameter = parameter
        
        self.button = ttk.Button(self, text = self.name, command = lambda: self.func(self, self.parameter))
        self.button.place(relx = 0, rely = 0, relwidth = 1, relheight = 1)
```

```python
class StartPage(Frame):

    def __init__(self, parent, controller):
        Frame.__init__(self,parent, bg="sea green")
        
        label = Label(self, text="Start Page", \
                font=12, bg="sea green", bd = 1, relief = SOLID)
        label.place(relx = 0, rely = 0, relwidth = 1)

        #####################################################  
        filename = labels_and_entry(self,"Filename: ","Molecules", lx = 0.1, ex = 0.22, ly = 0.1, ey = 0.1)
        temp = labels_and_entry(self, "Temp: ", 4, lx = 0.1, ex = 0.22, ly = 0.2, ey = 0.2, e_font = "Times 15 bold")
            
        entry1 = entry(self, "Numbers here", ex = 0.1, ey = 0.5, e_rw = 0.2)      
        button1 = buttons(self, "Button", func, entry1, bx = 0.1, by = 0.3)

        
root = FELion()
root.mainloop()
```

```python
def func(self, arg):
    self.arg = arg
    print(self.arg)
```

```python
def labels_and_entry(self, *args, **kwargs):
    label, entries = [i for i in args]    
    v = {'lx': float(), 'ly': float(), 'ex': float(), 'ey':float(), 'l_rh': 0.05, 'l_rw': 0.1,
                 'e_rw': 0.1, 'e_rh': 0.05, 'l_font': 'Times 12 bold', 'e_font': 'Times 12 italic'}
    
    for i in kwargs:
        if i in v:
            v[i] = kwargs[i]
            
    self.l1 = labels(self, label, lx = v['lx'], ly = v['ly'], l_rh = v['l_rh'], l_rw = v['l_rw'], l_font = v['l_font'])
    self.e1 = entry(self, entries, ex = v['ex'], ey = v['ey'], e_rh = v['e_rh'], e_rw = v['e_rw'], e_font = v['e_font'])

def buttons(self, name, func, parameter, **kwargs):
    
    v = {'bx': float(), 'by': float(),'b_rh': 0.05, 'b_rw': 0.1}
    
    for i in kwargs:
        if i in v:
            v[i] = kwargs[i]

    self.button = my_button(self, name, func, parameter)
    self.button.place(relx = v['bx'], rely = v['by'], relheight = v['b_rh'], relwidth = v['b_rw'])
    
def labels(self, label, **kwargs):
    
    v = {'lx': float(), 'ly': float(),'l_rh': 0.05, 'l_rw': 0.1, 'l_font':"Times 12 bold"}
    
    for i in kwargs:
        if i in v:
            v[i] = kwargs[i]
            
    self.label = my_label(self, label, v['l_font'])
    self.label.place(relx = v['lx'], rely = v['ly'], relheight = v['l_rh'], relwidth = v['l_rw'])

def entry(self, entries, **kwargs):
    
    v = {'ex': float(), 'ey':float(),'e_rh': 0.05, 'e_rw': 0.1, 'e_font':"Times 12 italic"}
    
    for i in kwargs:
        if i in v:
            v[i] = kwargs[i]

    self.entry = my_entry(self, entries, v['e_font'])
    self.entry.place(relx = v['ex'], rely = v['ey'], relheight = v['e_rh'], relwidth = v['e_rw'])
    self.value = self.entry.get()
    
    return self.value
```

```python
class FELion(Tk):
    def __init__(self, *args, **kwargs):

        Tk.__init__(self, *args, **kwargs)
        Tk.iconbitmap(self,default='C:/FELion-GUI/FELion_Icon.ico')
        Tk.wm_title(self, "FELion-Spectrum Analyser v.2.0")
        Tk.wm_geometry(self, "900x600")
       
        container = Frame(self)
        container.pack(side="top", fill="both", expand = True)
        container.grid_rowconfigure(0, weight=1)
        container.grid_columnconfigure(0, weight=1)

        StatusBarFrame = Frame(self)
        StatusBarFrame.pack(side = "bottom", fill = "both", expand = False)

        statusBar_left_text = "Version 2.0"
        statusBar_left = Label(StatusBarFrame)
        statusBar_left.config(text = statusBar_left_text, \
        relief = SUNKEN, bd = 2, font = "Times 10 italic", pady = 5, anchor = "w")
        statusBar_left.pack(side = "top", fill = "both", expand = True)

        statusBar_right_text = "Developed at dr. Sandra's Lab FELIX"
        statusBar_right = Label(StatusBarFrame)
        statusBar_right.config(text = statusBar_right_text, \
        relief = SUNKEN, bd = 2, font = "Times 10 italic", pady = 5, anchor = "e")
        statusBar_right.pack(side = "top", fill = "both", expand = True)

        self.frames = {}
        frame = StartPage(container, self)
        self.frames[StartPage] = frame
        frame.grid(row=0, column=0, sticky="nsew")

        self.show_frame(StartPage)

    def show_frame(self, cont):

        frame = self.frames[cont]
        frame.tkraise()
```
