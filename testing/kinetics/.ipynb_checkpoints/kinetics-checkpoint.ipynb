```python
import numpy as np
from scipy.optimize import curve_fit
from scipy.integrate import odeint
from uncertainties import ufloat
import matplotlib.pyplot as plt
```

# Compute rate constant 'k' while solving ODE rate equation dCa/dt

## Time and Concentration (Ca)

```python
# given data we want to fit
time = [0, 0.1, 0.2, 0.4, 0.8, 1]
Ca = [2.0081,  1.5512,  1.1903,  0.7160,  0.2562,  0.1495]
```

## Function to fit 'k' while solving the ODE

```python
def fitfunc(t, k):
    'Function that returns Ca computed from an ODE'
    
    def myode(Ca, t):
        'Function to fit k value for ODE equation'
        
        dCa_dt = -k * Ca # ODE equation to fit
        return dCa_dt

    Ca0 = Ca[0]
    Ca_soln = odeint(myode, Ca0, t) # Solving the ODE equation
    return Ca_soln[:,0]

k_fit, kcov = curve_fit(fitfunc, time, Ca)
k_err = np.diag(np.sqrt(kcov))
k = ufloat(k_fit[0], kcov[0])

time_fit = np.linspace(0,1);
Ca_fit = fitfunc(time_fit, k_fit)
```

## Plotting

```python
plt.figure(figsize = (15, 5), dpi = 150)
plt.plot(time, Ca, 'ro', label='data')
plt.plot(time_fit, Ca_fit, 'b.-', label=f'fit: k = {k:.4fP} /s')
plt.legend(loc='best')
plt.xlabel('Time (s)')
plt.ylabel('Ca')
plt.title('Solving rate eq from ODE: $dCa/dt = -k*Ca$ with computed k value')
plt.grid()
plt.show()
plt.close()
```
