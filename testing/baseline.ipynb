Baseline Correction using rampy

```python
import matplotlib.pyplot as plt
import numpy as np
import rampy as rp

%matplotlib inline
```

```python
filename = "./_datas/felixfile.felix"
spectrum = np.genfromtxt(filename)

base = np.genfromtxt('./_datas/basefile.base')
print(base[:,0])

bir = np.array([[750., 920., 1300.]]) # the frequency regions devoid of signal, used by rp.baseline()

y_corrected, background = rp.baseline(spectrum[:,0],spectrum[:,2],bir,"arPLS",lam=10e11)

plt.figure(figsize=(15,5), dpi = 100)

plt.plot(spectrum[:,0],spectrum[:,2],label="raw data")
plt.plot(spectrum[:,0],background,label="background")
plt.plot(spectrum[:,0],y_corrected,label="corrected signal")

plt.legend()
#plt.gca().invert_yaxis()
#print("Signal centroid is {}".format(rp.centroid(spectrum[:,0],y_corrected)))
```

```python
felix = np.genfromtxt('./_datas/felixdata.dat')
x, y = felix[:,0], felix[:,1]

bir = np.array([[750., 920.]])
y_corrected, background = rp.baseline(x, y,bir,"arPLS",lam=10e10)
original = plt.plot(x, y, label = 'original')
bkg = plt.plot(x, background, label = 'background')
correc = plt.plot(x, y_corrected, label = 'corrected')
plt.legend()
#plt.gca().invert_yaxis()
```
